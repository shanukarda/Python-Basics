```python
# How to handle missing values in pandas
import pandas as pd
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.tail()

```

```python
ufo.isnull().tail()
# isnull() shows a False value if something is not null
# if something is null then it shows True.
```

```python
# Exact opposite of isnull()

ufo.notnull().tail()
# isnull() shows a True value if something is not null
# if something is null then it shows False.
```

```python
ufo.isnull().sum()
```

```python
ufo[ufo.City.isnull()]
```

```python
ufo.shape
```

```python
#how to drop missing values rows

ufo.dropna(how='any').shape
```

```python
# It says drop the row if all column values are missing in that row,
# but we dont have any that type of dataset
ufo.dropna(how='all').shape
```

```python
# It says drop the row if city and shape reported column values are missing.
ufo.dropna(subset=['City','Shape Reported'],how='any').shape
```

```python
#by default missing values are excluded in value_counts()
ufo['Shape Reported'].value_counts()
```

```python

ufo['Shape Reported'].value_counts(dropna=False)
```

```python
# how to fill missing values in a column

ufo['Shape Reported'].fillna(value="Various",inplace = True)
```

```python
ufo["Shape Reported"].value_counts(dropna=False)
```

```python
#When should we use the 'inplace' parameter in pandas

import pandas as pd
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.head()
```

```python
ufo.shape
```

```python
ufo.drop('City',axis=1).head() # drop used to drop the column or row
```

```python
ufo.head() #but still City is not gone.

# Beacuse drop has 'inplace' parameter and if it is False then
# changes would not affect to main data.
```

```python
ufo.drop('City',axis=1,inplace=True)
#If want to change main data then set inplace =True
```

```python
# Now 'City' column will be gone
ufo.head()
```
